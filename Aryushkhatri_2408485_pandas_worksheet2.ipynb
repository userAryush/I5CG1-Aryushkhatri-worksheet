**Problem 2 - Data Imputations:**

```python
df2 = pd.read_csv('/content/drive/My Drive/Datasets/Datasets/medical_students_dataset.csv')
pd.DataFrame(df2)
```

2. Check info of the DataFrame and identify column with missing (null) values.

```python
print("      Information of second dataset:\n")
print(df2.info())

print("\nColumns with null values: ")

print(df2.isnull().sum())
```

3. For the column with missing values fill the values using various techniques we discussed above. Try
to explain why did you select the particular methods for particular column.

```python
# Use the median if the column is skewed or contains outliers that could distort the mean (i.e., the data has extreme values that are far from the rest).
df2['Age'] = df2['Age'].fillna(df2['Age'].median())

df2['BMI'] = df2['BMI'].fillna(df2['BMI'].median())

df2['Blood Pressure'] = df2['Blood Pressure'].fillna(df2['Blood Pressure'].median())


#Use the mean if the column has a normal distribution or is not heavily affected by outliers.
# eg. height weight and temperature
df2['Height'] = df2['Height'].fillna(df2['Height'].mean())

df2['Weight'] = df2['Weight'].fillna(df2['Weight'].mean())

df2['Temperature'] = df2['Temperature'].fillna(df2['Temperature'].mean())

df2['Heart Rate'] = df2['Heart Rate'].fillna(df2['Heart Rate'].mean())

df2['Cholesterol'] = df2['Cholesterol'].fillna(df2['Cholesterol'].mean())


#The mode ensures that the missing values are replaced with a valid and existing category from the dataset.
#In Gender the mode might be "Female" if females occur more frequently. Replacing missing values with "Female" maintains the overall structure of the data.
df2['Blood Type'] = df2['Blood Type'].fillna(df2['Blood Type'].mode()[0])  # Mode for categorical data

df2['Gender'] = df2['Gender'].fillna(df2['Gender'].mode()[0])


df2['Smoking'] = df2['Smoking'].fillna(df2['Smoking'].mode()[0])  # Mode for categorical data


df2['Diabetes'] = df2['Diabetes'].fillna(df2['Diabetes'].mode()[0])  # Mode for categorical data


# Use ffill() (Forward Fill) if the missing value can be reasonably assumed to be the same as the previous value.

df2['Student ID'] = df2['Student ID'].fillna(method='ffill')

# Check if there are any remaining missing values
print("\nRemaining missing values after filling:")
print(df2.isnull().sum())
```

4. Check for any duplicate values present in Dataset and do necessary to manage the duplicate items.
{Hint: dataset.duplicated.sum()}

```python
# Check for duplicates
duplicate_count = df2.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_count}")

# Drop duplicate rows if any exist
df2_cleaned = df2.drop_duplicates()

# Display the cleaned DataFrame
print("\nDataFrame after removing duplicates:")
print(df2_cleaned.head())
```

Problem - 1:
Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
median, maximum value, and minimum value of the ’Fare’ column?

```python
subset_df = df3[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]

# Retainig rows where Pclass==1
first_class_df = subset_df[subset_df['Pclass'] == 1]

# statistics for Fare
mean_fare = first_class_df['Fare'].mean()
median_fare = first_class_df['Fare'].median()
max_fare = first_class_df['Fare'].max()
min_fare = first_class_df['Fare'].min()

print(f"Mean Fare: {mean_fare}")
print(f"Median Fare: {median_fare}")
print(f"Maximum Fare: {max_fare}")
print(f"Minimum Fare: {min_fare}")
```

Problem2: How many null values are contained in the ’Age’ column in your subsetted DataFrame? Once you’ve found
this out, drop them from your DataFrame.

```python
print("Total null values in Age: ",first_class_df['Age'].isnull().sum())

# The dropna() function in pandas is used to remove rows that contain missing or null values.
first_class_df_cleaned = first_class_df.dropna(subset=['Age'])

# df.shape[0] gives the number of rows
print(f"Number of rows after dropping nulls: {first_class_df_cleaned.shape[0]}")
```

Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual-
ization to show how the survival distributions vary by gender.

```python
import matplotlib.pyplot as plt

mean_survival_by_gender = df3.groupby('Sex')['Survived'].mean()

print(mean_survival_by_gender)

plt.figure(figsize=(6, 4))
plt.bar(mean_survival_by_gender.index, mean_survival_by_gender.values, color=['pink', 'skyblue'])
plt.title('Mean Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Survival Rate')
plt.show()
```
